```python
response = rq.get(URL)
print(response.status_code)
```

```python
soup = BeautifulSoup(response.text, 'html.parser')
table = soup.find('table', 'GirdTable')
```

```python
table.text
```

```python
headers = []
for th in table.find("tr"):
    headers.append(th.text.strip())
    print(th.text.strip())
```

```python
headers = list(filter(None, headers))
```

```python
VN30INDEX = pd.DataFrame(columns=headers)
VN30INDEX
```

```python
rows = []
for tr in table.find_all('tr')[1:]:
    cell = []
    tds = tr.find_all("td")
    for td in tds:
        cell.append(td.text.strip())
    rows.append(cell)

pd.DataFrame(rows, columns=headers)
```
