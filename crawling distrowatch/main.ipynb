```python
from bs4 import BeautifulSoup
import requests
import time
```

```python
start_time = time.perf_counter()
```

```python
host_name = 'https://distrowatch.com/'
base_url = 'https://distrowatch.com/index.php?dataspan={}'
params = ['votes']
```

```python
url = base_url.format(params[0])
url
```

```python
req = requests.get(url)
req.status_code
```

```python
soup = BeautifulSoup(req.text, 'lxml')
```

```python
distribution_selector = 'td.phr2'
rating_selector = 'td.phr3'
```

```python
soup.select(distribution_selector)
```

```python
def processing_distribution(tag):
    data = dict(
        name=tag.get_text(),
        href=host_name + tag.a.get('href', None)
    )
    return data

distribution_name = list(map(processing_distribution, soup.select(distribution_selector)))
distribution_name
```

```python
soup.select(rating_selector)
```

```python
rating = list(map(lambda x: float(x.get_text()), soup.select(rating_selector)))
rating
```

```python
import pandas as pd
```

```python
distribution_name
```

```python
df = pd.DataFrame(distribution_name)
df
```

```python
df['rating'] = rating
```

```python
df
```

```python
def processing_li(tag):
    name = tag.b.get_text()[:-1]
    child = tag.select('a') + tag.select('font')
    
    def processing_a(child_tag):
        data = dict(name=child_tag.get_text())
        href = child_tag.get('href', None)
        if href:
            data['href'] = host_name + href
        return data
        
    child = list(map(processing_a, child))
    
    ret_data = dict(
        name=name,
        child=child
    )
    return ret_data
```

```python
from pprint import pprint
```

```python
def processing_one_distro(item):
    name, url, rating = item
    
    req = requests.get(url)
    soup = BeautifulSoup(req.text, 'lxml')
    
    td = soup.select_one('td.TablesTitle')
    
    li = td.select('li')
    
    distro_data = list(map(processing_li, li))
    
    MAGIC_NUMBER = 6
    distro_description = td.get_text().split('\n')[MAGIC_NUMBER]
    
    dat = dict(
        distro_name=name,
        distro_rating=rating,
        distro_data=distro_data,
        distro_description=distro_description
    )
    
    return dat
```

```python
data = list(zip(df['name'], df['href'], df['rating']))
data
```

```python
import concurrent.futures as cf
```

```python
with cf.ThreadPoolExecutor() as exe:
    final_data = list(exe.map(processing_one_distro, data))
```

```python
final_data
```

```python
import json
```

```python
json.dump(final_data, open('distrowatch_data.json', 'w'), indent=2)
```
